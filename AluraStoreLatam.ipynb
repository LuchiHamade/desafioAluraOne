### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

columnas_a_eliminar = ['lat', 'lon']

tienda = tienda.drop(columns=columnas_a_eliminar)
tienda2 = tienda2.drop(columns=columnas_a_eliminar)
tienda3 = tienda3.drop(columns=columnas_a_eliminar)
tienda4 = tienda4.drop(columns=columnas_a_eliminar)

#tienda.head()

# Poder ver las cinco primeras filas de las tiendas para verificar que est√©n bien cargadas

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"üì¶ Tienda {i}:")
    display(df.head())
    print("\n")  # espacio entre tablas

```

#1. An√°lisis de facturaci√≥n



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular ingresos por tienda
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()

# Creaci√≥n diccionario ingresos por tienda
ingresos_dict = {
    "Tienda 1": ingreso_t1,
    "Tienda 2": ingreso_t2,
    "Tienda 3": ingreso_t3,
    "Tienda 4": ingreso_t4
}



df_ingresos = pd.DataFrame({
    "Tienda": ingresos_dict.keys(),
    "Ingreso Total (sin formato)": ingresos_dict.values()
})

# Crear nueva columna con un formato m√°s legible
df_ingresos["Ingreso Total"] = df_ingresos["Ingreso Total (sin formato)"].apply(lambda x: f"${x:,.2f}")

# Mostrar tabla de ingresos
display(df_ingresos[["Tienda", "Ingreso Total"]])


#Gr√°fico de torta
plt.figure(figsize=(8, 8))
plt.pie(
    df_ingresos["Ingreso Total (sin formato)"],
    labels=df_ingresos["Tienda"],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('rocket', n_colors=len(df_ingresos))
)
plt.title("Ingresos por Tienda")
plt.tight_layout()
plt.show()


```

```python

# Agregar columna 'Tienda' en cada DataFrame
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar
datos_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)


# Asegurarse de que la fecha est√© bien convertida
datos_tiendas['Fecha de Compra'] = pd.to_datetime(datos_tiendas['Fecha de Compra'], dayfirst=True)

# Crear columna "Mes"
datos_tiendas['Mes'] = datos_tiendas['Fecha de Compra'].dt.to_period('M').astype(str)

# Calcular ingresos por fila
datos_tiendas['Ingreso'] = datos_tiendas['Precio'] + datos_tiendas['Costo de env√≠o']

# Agrupar ingresos por Mes y Tienda
ingresos_por_tienda = datos_tiendas.groupby(['Mes', 'Tienda'])['Ingreso'].sum().reset_index()

# Gr√°fico de l√≠neas por tienda
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=ingresos_por_tienda,
    x='Mes',
    y='Ingreso',
    hue='Tienda',  # <- cambiamos esto
    marker='o',
    palette='rocket'
)
plt.xticks(rotation=45)
plt.title('Ingresos Totales por Tienda (Mes a Mes)')
plt.xlabel('Mes')
plt.ylabel('Ingresos ($)')
plt.tight_layout()
plt.show()


```

## 2. Ventas por categor√≠a





```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Agregar columna 'tienda' a cada DataFrame individual
tienda["tienda"] = "Tienda_1"
tienda2["tienda"] = "Tienda_2"
tienda3["tienda"] = "Tienda_3"
tienda4["tienda"] = "Tienda_4"

# 2. Combinar DataFrames
datos_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# 3. Limpiar nombres columnas (min√∫sculas y guiones bajos)
datos_tiendas.columns = datos_tiendas.columns.str.strip().str.lower().str.replace(" ", "_")

# 4. Eliminar columnas duplicadas, dejando solo una 'tienda'
datos_tiendas = datos_tiendas.loc[:, ~datos_tiendas.columns.duplicated()]

# 5. Forzar columnas a string simple para evitar problemas
datos_tiendas['tienda'] = datos_tiendas['tienda'].astype(str)
datos_tiendas['categor√≠a_del_producto'] = datos_tiendas['categor√≠a_del_producto'].astype(str)

# 6. Agrupar por tienda y categor√≠a, contando ventas
ventas_por_categoria = datos_tiendas.groupby(['tienda', 'categor√≠a_del_producto']).size().reset_index(name='cantidad_ventas')

# 7. Ordenar globalmente por cantidad de ventas y tomar top 10 filas
top_10_global = ventas_por_categoria.sort_values('cantidad_ventas', ascending=False).head(10).reset_index(drop=True)

# Mostrar tabla sin √≠ndice
display(top_10_global.style.hide(axis="index"))

# Gr√°fico de barras horizontales ordenado
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_10_global,
    y='categor√≠a_del_producto',
    x='cantidad_ventas',
    hue='tienda',
    dodge=True,
    palette='rocket'
)
plt.title('Top 10 Categor√≠as M√°s Vendidas')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Categor√≠a del Producto')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()

```

# 3. Calificaci√≥n promedio de la tienda


```python

# Calcular el promedio de calificaci√≥n por tienda
calificaciones_promedio = (
    datos_tiendas
    .groupby("tienda")["calificaci√≥n"]
    .mean()
    .reset_index(name="calificaci√≥n_promedio")
    .sort_values("calificaci√≥n_promedio", ascending=False)
)

#  Formatear el promedio con 2 decimales
calificaciones_promedio["calificaci√≥n_promedio"] = calificaciones_promedio["calificaci√≥n_promedio"].round(2)

# Mostrar resultado
display(calificaciones_promedio)
```

# 4. Productos m√°s y menos vendidos

```python
# Asumimos que cada fila representa 1 unidad vendida
df['Unidades'] = 1

# Agrupamos por tienda y producto, contando unidades vendidas
ventas = df.groupby(['Lugar de Compra', 'Producto'])['Unidades'].sum().reset_index()

# Producto m√°s vendido por tienda
top = ventas.loc[ventas.groupby('Lugar de Compra')['Unidades'].idxmax()]
top['Tipo'] = 'M√°s vendido'

# Producto menos vendido por tienda
bottom = ventas.loc[ventas.groupby('Lugar de Compra')['Unidades'].idxmin()]
bottom['Tipo'] = 'Menos vendido'

# Unimos ambos en un solo DataFrame
resumen = pd.concat([top, bottom])
```

```python
#visualizaci√≥n

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=resumen, x='Lugar de Compra', y='Unidades', hue='Tipo', palette='coolwarm')
plt.title('Productos m√°s y menos vendidos por tienda')
plt.xlabel('Tienda')
plt.ylabel('Unidades vendidas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
