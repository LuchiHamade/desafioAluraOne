### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#tienda.head()

# Poder ver las cinco primeras filas de las tiendas para verificar que est√©n bien cargadas

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"üì¶ Tienda {i}:")
    display(df.head())
    print("\n")  # espacio entre tablas






```

#1. An√°lisis de facturaci√≥n



```python
# Calcular ingresos por tienda
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()

# Creaci√≥n diccionario ingresos por tienda
ingresos_dict = {
    "Tienda 1": ingreso_t1,
    "Tienda 2": ingreso_t2,
    "Tienda 3": ingreso_t3,
    "Tienda 4": ingreso_t4
}


# Mostrar resultados
print(f" Ingreso total Tienda 1: ${ingreso_t1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_t2:,.2f}")
print(f" Ingreso total Tienda 3: ${ingreso_t3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_t4:,.2f}")



df_ingresos = pd.DataFrame({
    "Tienda": ingresos_dict.keys(),
    "Ingreso Total (sin formato)": ingresos_dict.values()
})

# Crear nueva columna con un formato m√°s legible
df_ingresos["Ingreso Total"] = df_ingresos["Ingreso Total (sin formato)"].apply(lambda x: f"${x:,.2f}")

# Mostrar tabla de ingresos
display(df_ingresos[["Tienda", "Ingreso Total"]])


```

# 2. Ventas por categor√≠a

```python

# Agregar la columna 'tienda' a cada DataFrame individual
tienda["tienda"] = "Tienda_1"
tienda2["tienda"] = "Tienda_2"
tienda3["tienda"] = "Tienda_3"
tienda4["tienda"] = "Tienda_4"


# Combinar
datos_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Limpiar columnas
datos_tiendas.columns = datos_tiendas.columns.str.strip().str.lower().str.replace(" ", "_")

# Imprimir la columna tienda
print(datos_tiendas["tienda"].head())

# Agrupar por tienda y categor√≠a
ventas_por_categoria = datos_tiendas.groupby(['tienda', 'categor√≠a_del_producto']).size().reset_index(name='cantidad_ventas')

# 6. Ordenar para ver las categor√≠as m√°s populares por tienda (opcional)
ventas_por_categoria = ventas_por_categoria.sort_values(['tienda', 'cantidad_ventas'], ascending=[True, False])

# 7. Mostrar resultado
display(ventas_por_categoria)
```

# 3. Calificaci√≥n promedio de la tienda


```python

# Calcular el promedio de calificaci√≥n por tienda
calificaciones_promedio = (
    datos_tiendas
    .groupby("tienda")["calificaci√≥n"]
    .mean()
    .reset_index(name="calificaci√≥n_promedio")
    .sort_values("calificaci√≥n_promedio", ascending=False)
)

# ‚úÖ Formatear el promedio con 2 decimales
calificaciones_promedio["calificaci√≥n_promedio"] = calificaciones_promedio["calificaci√≥n_promedio"].round(2)

# ‚úÖ Mostrar resultado
display(calificaciones_promedio)
```
